# Librosa: audio2plot - MP3, WAV

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt

filename = test_case_1

# Load the audio as a waveform 'y'
# Store the sampling rate as 'sr'
audio, sr = librosa.load(filename)

# Time range
time = np.linspace(0, audio.shape[0]/sr, num=audio.shape[0])

plt.plot(time, audio, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (quantized)')
plt.title('MP3 file visualization')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

# SpeechRecognition: speech2text - WAV

```python
import speech_recognition as sr

audio_filename = test_case_1

my_audio = sr.AudioFile(audio_filename)
r = sr.Recognizer()
with my_audio as source:
  audio = r.record(source)

your_speech = r.recognize_google(audio, language='en-US')
print(f'Audio transcription: {your_speech}')
```

# gtts: text2speech - MP3

```python
test_case_2 = "Tôi yêu AIO"

from gtts import gTTS
import librosa
import numpy as np

lang='en'
output_filename = 'record-gtts.mp3'
content = test_case_2

# text to speech
output = gTTS(content, lang=lang, slow=False)
output.save(output_filename)

 # load output audio
data, sr = librosa.load(output_filename)

# display audio
import IPython
IPython.display.display(IPython.display.Audio(np.transpose(data), rate=sr))
```
