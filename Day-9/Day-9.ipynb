```python
lst_data = []
while (len(lst_data) < 10):
  num = random.randint(1, 10)
  if num not in lst_data:
    lst_data.append(num)
print('List: ',lst_data)

lst_data2 = sorted(lst_data)
print('Cau 1:', lst_data2)

n = len(lst_data2)
if n%2 != 0:
  indx = int((n+1)/2)-1
  m = lst_data2[indx]
else:
  indx = int(n/2)-1
  m = (lst_data2[indx] + lst_data2[indx+1]) /2
print('Cau 2: Median:',m)

lst_odd_filter = []
for number in lst_data:
  if number%2 != 0:
    lst_odd_filter.append(number)
lst_odd_filter = sorted(lst_odd_filter,reverse=True)
print('Cau 3:',lst_odd_filter)
```

```python
# Random

import random
length = int(round(10*random.random(),0))

lst_data = []
while (len(lst_data) < length):
  num = random.randint(1, 10)
  lst_data.append(num)
print(f'List: {lst_data}. Length: {length}')

lst_data2 = sorted(lst_data)
print('Cau 1:', lst_data2)

n = len(lst_data2)
if n%2 != 0:
  indx = int((n+1)/2)-1
  m = lst_data2[indx]
else:
  indx = int(n/2)-1
  m = (lst_data2[indx] + lst_data2[indx+1]) /2
print('Cau 2: Median:',m)

lst_odd_filter = []
for number in lst_data:
  if number%2 != 0:
    lst_odd_filter.append(number)
lst_odd_filter = sorted(lst_odd_filter,reverse=True)
print('Cau 3:',lst_odd_filter)
```
