```python
text = 'Công cụ Suno AI nhanh chóng nhận được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày.'
from translate import Translator
translator_en = Translator(from_lang='vi', to_lang='en')
translation_en = translator_en.translate(text)
translator_jp = Translator(from_lang='vi', to_lang='ja')
translation_jp = translator_jp.translate(text)
print('Thu vien Translate:')
print(f'+ Tieng Anh: {translation_en}')
print(f'+ Tieng Nhat: {translation_jp}')

from googletrans import Translator
translate = Translator()
result_en = translate.translate(text, dest = 'en')
result_jp = translate.translate(text, dest = 'ja')
print('Thu vien Googletrans:')
print(f'+ Tieng Anh: {result_en.text}')
print(f'+ Tieng Nhat: {result_jp.text}')
```
