```python
import numpy as np
import torch
import tensorflow as tf
import random
```

## 1D

```python
lst = []
length = random.randint(5,15)
length
```

```python
while len(lst) < length:
  num = random.randint(1,10)
  lst.append(num)
lst
```

```python
arr_1D = np.array(lst)
arr_1D
```

```python
tensor_1D_pt = torch.tensor(lst)
tensor_1D_pt
```

```python
tensor_1D_tf = tf.constant(lst)
tensor_1D_tf
```

## 2D

```python
lst_2D = []
length = random.randint(2,4)
sub_lst = []
for j in range(length):
  for i in range (length):
    sub_lst.append(random.randint(1,10))
  lst_2D.append(sub_lst)
  sub_lst = []
lst_2D
```

```python
arr_2D = np.array(lst_2D)
arr_2D
```

```python
tensor_2D_pt = torch.tensor(lst_2D)
tensor_2D_pt
```

```python
tensor_2D_tf = tf.convert_to_tensor(lst_2D)
tensor_2D_tf
```

```python
print(arr_2D.shape)
print(arr_2D.dtype)
print(type(arr_2D))
```

```python
print(tensor_2D_pt.shape)
print(tensor_2D_pt.dtype)
print(tensor_2D_pt.device)
print(type(tensor_2D_pt))
```
