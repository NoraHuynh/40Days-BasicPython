```python
# Cau 1
lst_data = []
for num in range (2,13,2):
  lst_data.append(num)
print(f'Cau 1: {lst_data}')

# Cau 2:
for num in lst_data:
  if num % 3 == 0:
    lst_data.remove(num)
print(f'Cau 2: {lst_data}')

# Cau 3:
for i in range (1,4):
  lst_data.append(i)
lst_num = [6,7,8]
for i in range (len(lst_num)):
  lst_data.insert(3+i,lst_num[i])
print(f'Cau 3: {lst_data}')

# Cau 4:
for i,num in enumerate(lst_data):
  if (num%2 == 0) or (num%5 == 0):
    lst_data[i] = 0
print(f'Cau 4: {lst_data}')
```
