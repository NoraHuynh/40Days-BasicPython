```python
def lst_none(lst_data):
  lst = [indx for indx in range(len(lst_data)) if lst_data[indx] == None]
  return lst

def nearest_neighbor(lst_data):
  none_lst = lst_none(lst_data)
  nearest_distance = float('inf')
  for i in none_lst:
    for indx in range(len(lst_data)):
      if (indx not in none_lst):
        if (abs(i-indx) < nearest_distance):
          nearest_distance = abs(i-indx)
          nearest_val = indx
    lst_data[i] = lst_data[nearest_val]
    nearest_distance = float('inf')
  return lst_data

# Input
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
print(f'Before: {lst_data}')
# Output
print(f'After: {nearest_neighbor(lst_data)}')
```
