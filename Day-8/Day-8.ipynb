```python
lst_name = [['Ron','Harry','Hermione'],
            ['Albus','Marcus','Draco'],
            ['Newt','Susan','Cedric'],
            ['Luna','Penelope','ChoChang']]
gryffindor_studnts = ['Ron','Harry','Hermione']
studnt_name = ['Harry', 'Marcus','Ron','Luna']
for i, row in enumerate(lst_name):
  for j,studnt in enumerate(row):
    for st_n in studnt_name:
      if studnt == st_n:
        if studnt in gryffindor_studnts:
          print(f'{st_n}\'s in Gryffindor\'s house.')
          print(f'Student ID: {i+1}.{j+1}.')
        else:
          print(f'{st_n}\'s NOT IN Gryffindor\'s house.')
          print(f'Student ID: {i+1}.{j+1}.')
```
