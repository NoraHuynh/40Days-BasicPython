```python
import numpy as np
import tensorflow as tf
import torch
```

```python
# Cau 1:
arr_np = np.arange(0,10,1)
arr_tf = tf.range(0,10,1)
arr_pt = torch.arange(0,10,1)

print('Cau 1:')
print(f'- Numpy array: {arr_np}')
print(f'- Tensorflow tensor: {arr_tf}')
print(f'- Pytorch tensor: {arr_pt}')
```

```python
# Cau 2:
matrx_np = np.eye(3,3)
matrx_tf = tf.eye(3,3)
matrx_pt = torch.eye(3,3)

print('Cau 2:')
print(f'- Numpy matrix:\n {matrx_np}')
print(f'- Tensorflow matrix:\n {matrx_tf}')
print(f'- Pytorch matrix:\n {matrx_pt}')
```

```python
# Cau 3:
np.random.seed(2024)
ran_np = np.random.rand(3,4)
randint_np = np.random.randint(-10,10,(3,4))

torch.manual_seed(2024)
ran_pt = torch.rand(3,4)
randint_pt = torch.randint(-10,10,(3,4))

tf.random.set_seed(2024)
ran_tf = tf.random.uniform((3,4),minval=0, maxval=1)
randint_tf = tf.random.uniform((3,4),minval=-10, maxval=10,dtype='int32')

print('Cau 3:')
print('* Tensor trong khoang [0,1]:')
print(f'- Numpy tensor:\n {ran_np}')
print(f'- Pytorch tensor:\n {ran_pt}')
print(f'- Tensorflow tensor:\n {ran_tf}')
print('')
print('* Tensor trong khoang [-10,10]:')
print(f'- Numpy tensor:\n {randint_np}')
print(f'- Pytorch tensor:\n {randint_pt}')
print(f'- Tensorflow tensor:\n {randint_tf}')
```
