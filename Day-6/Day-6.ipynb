```python
#1. If-else
time = input('Input time: ')
if time == '5 AM':
  print('Wake up')
elif time == '6 AM':
  print('Yoga')
elif time == '7 AM':
  print('Work')
else:
  print('Do something else')
```

```python
#2. Dictionary
time = input('Input time: ')
lst = {'5 AM':'Wake up',
       '6 AM':'Yoga',
       '7 AM':'Work'}
lst.get(time,'Do something else')
```

```python
#3. match-case
time = input('Input time: ')
match time:
  case '5 AM':
    print('Wake up')
  case '6 AM':
    print('Yoga')
  case '7 AM':
    print('Work')
  case _:
    print('Do something else')
```
