```python
lst_data = [1, 1.1, None, 1.4, None, 1.5, None, 2.0]
indx_lst = []
for i in range(len(lst_data)):
  if lst_data[i] == None:
    indx_lst.append(i)
print(f'Vi tri None dau tien: {indx_lst[0]}')
print(f'Danh sach vi tri co gia tri None: {indx_lst}')
```
