```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# Cau 1:
print(np.zeros((3,4)))
print(torch.zeros((3,4)))
print(tf.zeros((3,4)))
```

```python
# Cau 2:
print(np.ones((3,4)))
print(torch.ones((3,4)))
print(tf.ones((3,4)))
```

```python
# Cau 3:
print(np.full((3,4),5))
print(torch.full((3,4),5))
print(tf.fill((3,4),5))
```
