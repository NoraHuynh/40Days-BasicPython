```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Cau 1:
img_path = '/content/drive/MyDrive/Colab Notebooks/40-Day Python/Day-26_Image_dataset/1.jpg'

# Read image
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Cau 3:
import numpy as np

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image")
plt.show()

bright_img = img.astype(np.float32) + 50
bright_img = np.clip(bright_img, 0, 255)
bright_img = bright_img.astype(np.uint8)
plt.imshow(bright_img)
plt.title('Brightness Image')
plt.show()

dark_img = img.astype(np.float32) - 80
dark_img = np.clip(dark_img, 0, 255)
dark_img = dark_img.astype(np.uint8)
plt.imshow(dark_img)
plt.title('Darkness Image')
plt.show()
```
