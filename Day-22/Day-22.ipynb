# While-Loop

```python
import random
def random_number_with_condition(total):
  random.seed(0)
  cnt_time = 0
  a = random.randint(1,20)
  b = random.randint(1,20)
  if a + b != total:
    cnt_time += 1
  while a+b != total:
    a = random.randint(1,20)
    b = random.randint(1,20)
    cnt_time += 1
  return cnt_time

print(f'- Test case 1: random_number_with_condition(40) -> {random_number_with_condition(40)} lan')
print(f'- Test case 2: random_number_with_condition(20) -> {random_number_with_condition(20)} lan')
print(f'- Test case 3: random_number_with_condition(35) -> {random_number_with_condition(35)} lan')
```

# While Loop + continue and break keywords

```python
def find_divisible_number(a):
  num = 101
  while num%a != 0:
    num += 1
  return num

print(f'- Test case 1: find_divisible_number(5) -> {find_divisible_number(5)}')
print(f'- Test case 2: find_divisible_number(17) -> {find_divisible_number(17)}')
```

# While-Loop Exercise

```python
def f(x,a):
  f_x = x**2 - a
  return f_x

def deviate_f(x):
  res = 2*x
  return res

def find_squared_root(a):
  x = [0]*1000
  EPSILON = 0.001
  x[0] = a
  n = 0
  x[n+1] = x[n] - f(x[n],a)/deviate_f(x[n])
  while abs(x[n+1] - x[n]) >= EPSILON:
    n += 1
    x[n+1] = x[n] - f(x[n],a)/deviate_f(x[n])
  return x[n+1]

print(f'- Test case 1: find_squared_root(2) -> {find_squared_root(2)}')
print(f'- Test case 2: find_squared_root(3) -> {find_squared_root(3)}')
```
