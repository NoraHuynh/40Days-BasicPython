# Create file

```python
# Cau 1:
lst_data = [str(i) for i in range(1,11)]
data_to_write = "-".join(lst_data)

with open('/content/drive/MyDrive/Colab Notebooks/40-Day Python/Day-19/data.txt', 'w') as f:
    f.write(data_to_write)
print(f'Cau 1: {data_to_write}')
```

```python
# Cau 2:
with open('/content/drive/MyDrive/Colab Notebooks/40-Day Python/Day-19/data.txt','r') as f:
  lst = f.read().split('-')

lst_filter = []
for item in lst:
  num = int(item)
  if num%3 == 0:
    lst_filter.append(num)

print(f'Cau 2: {lst_filter}')
```
