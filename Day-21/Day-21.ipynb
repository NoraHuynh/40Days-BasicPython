```python
# Cau 1:
# Read File
file_path = '/content/drive/MyDrive/Colab Notebooks/40-Day Python/Day-21/Day-21_FileAdvance_data.txt'
with open(file_path,'r') as f:
  data = f.read()

data = data.replace('\n',' ').lower()
print(f'Cau 1: {data}')
```

```python
# Cau 2:
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

tokenization = word_tokenize(data)
distinct_words = set(tokenization)

print((len(tokenization),len(distinct_words)))
print(f'Cau 2: {distinct_words}')
```

```python
# Cau 3:
cnt_dict = {}
for word in distinct_words:
  cnt_dict[word] = tokenization.count(word)

max_freq = max(cnt_dict,key=cnt_dict.get) # to get the max key
min_freq = min(cnt_dict,key=cnt_dict.get)

print('Cau 3:')
print(f'+ "thought" in data is {cnt_dict.get("thought", 0)}') # to get the frequency of "thought"
print(f'+ "you" in data is {cnt_dict.get("you", 0)}')
print(f'+ "{max_freq}" is the most frequent word, frequency: {cnt_dict[max_freq]}')
print(f'+ "{min_freq}" is the least frequent word, frequency: {cnt_dict[min_freq]}')
```
