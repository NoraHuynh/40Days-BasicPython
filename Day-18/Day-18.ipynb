# List Comprehension

```python
stop_words = ['I','love','and','to']
input = 'I love AI and listen to music'
print([word for word in input.split(' ') if word not in stop_words])
```

# List Tuple

```python
# Cau 1:
my_tuple1 = (2,3)
my_tuple2 = (3,6)

# Cau 2:
def cal_sum_mul(tuple_data):
  sum = 0
  mul = 1
  for i in range(len(tuple_data)):
    sum += tuple_data[i]
    mul *= tuple_data[i]
  return (sum,mul)
print(f'Cau 2: Result_vector1={cal_sum_mul(my_tuple1)}, Result_vector2={cal_sum_mul(my_tuple2)}')

# Cau 3:
import math

def cal_dist(vector1, vector2):
  sum_square = 0
  for j in range(len(vector1)):
    sum_square += (vector1[j]-vector2[j])**2
  res = math.sqrt(sum_square)
  return res
print(f'Cau 3: sqrt({round(cal_dist(my_tuple1,my_tuple2)**2)})={cal_dist(my_tuple1,my_tuple2)}')

# Cau 4:
print(f'Cau 4: index={(my_tuple1.index(3),my_tuple2.index(3))}')
```
